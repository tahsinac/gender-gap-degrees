## Visualizing The Gender Gap In College Degrees


__[The Department of Education Statistics](http://url)__ releases a data set annually containing the percentage of bachelor's degrees granted to women from 1970 to 2012. The data set is broken up into 17 categories of degrees, with each column as a separate category.

Randal Olson, a data scientist at University of Pennsylvania, has cleaned the data set and made it available on his personal website. The dataset can be found __[here](http://www.randalolson.com/wp-content/uploads/percent-bachelors-degrees-women-usa.csv)__.

The aim of this project is to explore the gender gap in STEM and other degrees.

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)

# Reading in the csv file
women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')

# List of degrees to be explored
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

# Generating a 6x3 grid of subplots
fig = plt.figure(figsize=(16, 20))

# Generating the first column of line charts: STEM degrees.
for i in range(0,18,3):
    index = int(i/3)
    ax = fig.add_subplot(6,3,i+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[index]], c=cb_orange, label='Men', linewidth=3)
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.axhline(50, c = (137/255,137/255,137/255), alpha=0.3, linestyle = '--')
    ax.set_title(stem_cats[index])
    ax.tick_params(bottom="off", top="off", left="off", right="off",labelbottom='off')
    if index == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif index == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
        ax.tick_params(labelbottom='on')

# Generating the second column of line charts: Liberal Arts degrees.
for i in range(1,16,3):
    index = int((i-1)/3)
    ax = fig.add_subplot(6,3,i+1)
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[index]], c=cb_orange, label='Men', linewidth=3)
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.axhline(50, c = (137/255,137/255,137/255), alpha=0.3, linestyle = '--')
    ax.set_title(lib_arts_cats[index])
    ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom='off')
    if index == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif index == 4:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
        ax.tick_params(labelbottom='on')

# Generating third column of line charts. Other degrees.
for i in range(2,20,3):
    index = int((i-2)/3)
    ax = fig.add_subplot(6,3,i+1)
    ax.plot(women_degrees['Year'], women_degrees[other_cats[index]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[index]], c=cb_orange, label='Men', linewidth=3)
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.axhline(50, c = (137/255,137/255,137/255), alpha=0.3, linestyle = '--')
    ax.set_title(other_cats[index])
    ax.tick_params(bottom="off", top="off", left="off", right="off", labelbottom='off')
    if index == 0:
        ax.text(2005, 91, 'Men')
        ax.text(2005, 6, 'Women')
    elif index == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2005, 25, 'Women')
        ax.tick_params(labelbottom='on')

plt.show

```

### Findings



* Engineering, by far, has the largest gender gap and this has only decreased slightly over 42 years.


* The gender gap in computer science was condirably large in 1970. However, this gap was closed down by quite a bit, only for it to crease back to it's initial value.


* Amongst the STEM degrees, psychology had the smallest gender gap in 1970, with woming being the slight majority. The number of degrees awarded to men and women were equal fo a small period of time. The gender gap then increased steadily until 2012 with men being the majority.
